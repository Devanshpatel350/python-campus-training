```python
# Q1. Write a Python program to create two sets, A and B, and find the elements that are common to both sets.
# A = {1, 2, 3, 4, 5}
# B = {4, 5, 6, 7, 8}

A = {1,2,3,4,5}
B = {4,5,6,7,8}
common_elements = A.intersection(B)
print("Common elements in sets A and B:", common_elements)
```

```python
# Q2. Write a Python program to create two sets, X and Y, and find the elements that are unique to each set (not present in both).
# X = {10, 20, 30, 40}
# Y = {30, 40, 50, 60}

X = {10, 20, 30, 40}
Y = {30, 40, 50, 60}
unique_to_X = X.difference(Y)
unique_to_Y = Y.difference(X)
print("Element unique to set X:", unique_to_X)
print("Element unique to set Y:", unique_to_Y)
```

```python
# Q3. Write a Python program to check if all elements of one set are contained within another set.
# S1 = {1, 2, 3}
# S2 = {1, 2, 3, 4, 5}

S1 = {1, 2, 3}
S2 = {1, 2, 3, 4, 5}
is_subset = S1.issubset(S2)
print("All elements of set S1 are contained in set S2:", is_subset)
```

```python
# Q4. Write a Python program which create a new set from given two sets have only that elements which are not common.
# A = {2, 4, 6}
# B = {1, 3, 5}

A = {2, 4, 6}
B = {1, 3, 5}
unique_elements = A.symmetric_difference(B) 
print(unique_elements)
```

```python
# Q5. Write a Python program to combine all elements from three different sets into one.
# Set1 = {1, 2}
# Set2 = {2, 3}
# Set3 = {3, 4}

Set1 = {1, 2}
Set2 = {2, 3}
Set3 = {3, 4}
combined_set = Set1.union(Set2).union(Set3)
print("Combined set of all elements from Set1, Set2, and Set3:", combined_set)
```

```python
# Q6. Write a Python program to find all elements that are in one set but not in another.
# A = {10, 20, 30}
# B = {20, 30, 40}
A = {10, 20, 30}
B = {20, 30, 40}
elements_in_A_not_in_B = A.difference(B)
elements_in_B_not_in_A = B.difference(A)
print("Element in A not in B ",elements_in_A_not_in_B)
print("Element in B not in A ",elements_in_B_not_in_A)
```

```python
# Q7. Write a Python program to remove a specific element(n) from a set, but only if it exists in the set.
# S = {1, 2, 3, 4, 5}
# Test case 1: n = 2
# Test case 2: n = 7 
S = {1, 2, 3, 4, 5}
n = 7
if n in S:
    S.remove(n)
    print(f"Element {n} removed from the set. Updated set: {S}")
else:
    print(f"Element {n} does not exist in the set. No changes made. Set: {S}")
```
